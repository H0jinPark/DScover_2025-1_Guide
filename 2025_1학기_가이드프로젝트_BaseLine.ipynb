### ë°ì´í„° ë¶ˆëŸ¬ì˜¤ê¸°

```python
import pandas as pd
import numpy as np

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score
```

```python
train = pd.read_csv('./train.csv')
```

## ë°ì´í„° ì „ì²˜ë¦¬

### ê°„ë‹¨í•œ ëª¨ë¸ë§ì— í•„ìš”í•œ ê°„ë‹¨í•œ ì „ì²˜ë¦¬ë¥¼ ìˆ˜í–‰í•˜ëŠ” ì½”ë“œì…ë‹ˆë‹¤

```python
x = train.drop(['ID','í—ˆìœ„ë§¤ë¬¼ì—¬ë¶€'],axis=1)
y = train['í—ˆìœ„ë§¤ë¬¼ì—¬ë¶€']
```

```python
# SimpleImputer : í‰ê·  ëŒ€ì²´
mean_imputer = SimpleImputer(strategy='mean')

# ê²°ì¸¡ê°’ì„ í‰ê· ìœ¼ë¡œ ëŒ€ì²´í•  ì—´ ëª©ë¡
columns_fill_mean = ['í•´ë‹¹ì¸µ', 'ì´ì¸µ','ì „ìš©ë©´ì ','ë°©ìˆ˜', 'ìš•ì‹¤ìˆ˜','ì´ì£¼ì°¨ëŒ€ìˆ˜']

# í•™ìŠµ ë°ì´í„°ì— fit ë° transform
x[columns_fill_mean] = mean_imputer.fit_transform(x[columns_fill_mean])
```

```python
# Label Encoding ì ìš© ì—´
label_encode_cols = ['ì¤‘ê°œì‚¬ë¬´ì†Œ','ê²Œì¬ì¼','ì œê³µí”Œë«í¼','ë°©í–¥']

label_encoders = {}
for col in label_encode_cols:
    le = LabelEncoder()
    x[col] = le.fit_transform(x[col].astype(str))
    label_encoders[col] = le
```

```python
# One-Hot Encoding ì ìš© ì—´
one_hot_cols = ['ë§¤ë¬¼í™•ì¸ë°©ì‹', 'ì£¼ì°¨ê°€ëŠ¥ì—¬ë¶€']

# One-Hot Encoding ì ìš©
one_hot_encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

# Train ë°ì´í„° ë³€í™˜
x_encoded = one_hot_encoder.fit_transform(x[one_hot_cols])
x_encoded_df = pd.DataFrame(x_encoded, columns=one_hot_encoder.get_feature_names_out(one_hot_cols), index=x.index)

# ê¸°ì¡´ ë°ì´í„°ì™€ ë³‘í•©
x = pd.concat([x.drop(columns=one_hot_cols), x_encoded_df], axis=1)
```

## í•™ìŠµí•˜ê¸°

```python
# ë°ì´í„° ë‚˜ëˆ„ê¸°
x_train, x_val, y_train, y_val = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
```

```python
# ì •í™•ë„, f1 score ê³„ì‚°
y_pred = model.predict(x_val)
y_pred_binary = (y_pred > 0.5).astype(int) # ì˜ˆì¸¡ì„ ìœ„í•´ì„œ ê°’ì„ binaryë¡œ ë³€ê²½!
accuracy = accuracy_score(y_val, y_pred_binary)
f1 = f1_score(y_val, y_pred_binary)
print(f"Accuracy: {accuracy}")
print(f"f1_score: {f1}")
```

### ì˜ˆì¸¡í•˜ê¸°

```python
# Test ë°ì´í„° ë¡œë“œ
test = pd.read_csv('./test.csv')
```

```python
# Test ê²°ì¸¡ê°’ ëŒ€ì²´
test[columns_fill_mean] = mean_imputer.transform(test[columns_fill_mean])
```

```python
# Label Encoding
for col in label_encode_cols:
    if col in test.columns:
        le = label_encoders[col]
        test[col] = test[col].astype(str)
        unseen = set(test[col].unique()) - set(le.classes_)
        if unseen:
            le.classes_ = np.append(le.classes_, list(unseen))
        test[col] = le.transform(test[col])
```

```python
# One-Hot Encoding
test_encoded = one_hot_encoder.transform(test[one_hot_cols])
test_encoded_df = pd.DataFrame(test_encoded, columns=one_hot_encoder.get_feature_names_out(one_hot_cols), index=test.index)

test = pd.concat([test.drop(columns=one_hot_cols), test_encoded_df], axis=1)
```

```python
test.drop(columns=['ID'],inplace=True)
```

```python
pred = pd.Series(model.predict(test))
```
